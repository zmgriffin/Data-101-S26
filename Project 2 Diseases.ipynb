```python
# Load libraries and data set
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportion_confint, proportions_ztest, proportion_effectsize
from scipy import stats
from statsmodels.stats.power import NormalIndPower, TTestIndPower, TTestPower
df = pd.read_csv("contagious_disease.csv")
```

```python
# use info and head to get a sense of the data set
df.head()
```

```python
# Filter for just pertussis (whooping cough) 
wc = df[(df['disease'] == 'Pertussis')]

# filter in the years 1947, 48, 
filter_years = [1947, 1948]

wc2 = wc[wc['year'].isin(filter_years)]

# filter for state is District of Columbia, Maryland or Virginia
filter_state = ['District Of Columbia', 'Maryland', 'Virginia']

wc3 = wc2[wc2['state'].isin(filter_state)]

wc3

```

```python
# Group by year and sum total cases of Whooping cough and US population
wc_counts = wc3.groupby('year')[['count','population']].sum()

wc_counts
```

### Difference Two Proportion Test: 
Ho: The proportion of whooping cough cases in the DMV before and after the DTP vaccine are equal.

Ha: The proportion of whooping cough cases in the DMV before and after the DPT vaccine are not equal.

```python
# set up variables 

#counts = ([wc_counts.at[1947, 'count'], [wc_counts.at[1948, 'count']]]) 
#nobs = ([wc_counts.at[1947, 'population'], [wc_counts.at[1948, 'population']]])
# I get errors about array size that I don't really understand when I try to access the counts and nobs from my table from the proportions test code,
# but it works fine when I manually type in the numbers 
counts = np.array([8558, 6038197])
nobs = np.array([3510, 6180523])

# run the difference of two proportion test
stat, pval = proportions_ztest(counts, nobs)
print(f"Z-statistic: {stat:.3f}, p-value: {pval:.4f}\n")

# interperet
if pval < 0.05:
    print("Reject Ho: Whooping cough proportions were different in the DMV before and after the DPT vaccine. ")
else:
    print("Fail to reject H0: no evidence that whooping cough proportions were different in the DMV before and after the DPT vaccine.")
```
