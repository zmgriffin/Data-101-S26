## Part 2)

### Imputting Embarked Missing Data

```python
# finding the most common value 
emb = ti['Embarked'].mode()[0] #most embarked location
```

```python
# from the previous info summary we know there are two NaN values for embarked
ti2 = ti.copy() # making a new dataframe in case I want to see the original for some reason
ti2['Embarked'] = ti2['Embarked'].fillna('emb')
ti2.info() # check to see if the non-null count matches the number of entries, yay it does
```

### Age: Mean Method

```python
allmean = ti2.copy()
allmean['Age'] = allmean['Age'].fillna(allmean['Age'].mean())
```

```python
allmean.info()
```

### Age: Median for Sex and Pclass

```python
#first_female.info() #see if there are NaN values
```

```python
first_female['Age'] = first_female['Age'].fillna(first_female['Age'].median())
first_female.head()
first_female.info() # non-null count to see check if it worked
```

```python
female = ti3[(ti3['Sex'] == 'female')].copy()
female.head()
male = ti3[(ti3['Sex'] == 'male')].copy()
male.head()
child = ti3[(ti3['Age'] < 16)].copy() # I an interpretting the assignment's question 'under sixteen' to not include 16 yearolds
child.head()
```

```python
status_f = female.groupby('Survived')['Survived'].agg('count')
alive_f = round(float(status_f[1])/len(female['Name']),4)
print(f'proportion of females who survived is {alive_f}')
```

```python
status_m = male.groupby('Survived')['Survived'].agg('count')
alive_m = round(float(status_m[1])/len(male['Name']),4)
print(f'proportion of males who survived is {alive_m}')
```

```python
status_ch = child.groupby('Survived')['Survived'].agg('count')
alive_ch = round(float(status_ch[1])/len(child['Name']),4)
print(f'proportion of children under 16 who survived is {alive_ch}')
```

```python
f1_life = first_female.groupby('Survived')['Survived'].agg('count')
f1_alive = round(float(f1_life[1])/len(first_female['Name']),4)
f1_alive

f2_life = sec_female.groupby('Survived')['Survived'].agg('count')
f2_alive = round(float(f2_life[1])/len(sec_female['Name']),4)
f2_alive

f3_life = third_female.groupby('Survived')['Survived'].agg('count')
f3_alive = round(float(f3_life[1])/len(third_female['Name']),4)
f3_alive
```

```python
m1_life = first_male.groupby('Survived')['Survived'].agg('count')
m1_alive = round(float(m1_life[1])/len(first_male['Name']),4)
m1_alive

m2_life = sec_male.groupby('Survived')['Survived'].agg('count')
m2_alive = round(float(m2_life[1])/len(sec_male['Name']),4)
m2_alive

m3_life = third_male.groupby('Survived')['Survived'].agg('count')
m3_alive = round(float(m3_life[1])/len(third_male['Name']),4)
m3_alive
```
