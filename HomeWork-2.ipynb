## Part 1) Working With Numpy: 

```python
#import numpy
import numpy as np
```

```python
# make a list of student names. Kept getting errors when I tried to make the empty list and then append this into it.
names = (input(f'Type Student Names, Separated by Coma: ').split(','))
print(names)
num_stu = int(len(names))
#print(num_stu)
```

```python
grades = np.empty((0,5),float)
print(grades)
 
```

```python
for i in range(num_stu):
    scores = input(f'type {names[i]} Homework, Project, Exam1, Exam2,, and Final Exam grades, separated by comma').split(',')
    scores = [float(scores[0]), float(scores[1]), float(scores[2]), float(scores[3]), float(scores[4])]
    new_row = np.array(scores)
    #print(new_row)
    grades = np.append(grades, [new_row], axis = 0)
print(grades)
  
```

```python
# weight vector
weight = np.array([.15,.15,.20,.20,.30])
print(weight)
```

```python
# calculate final grade via dot product
final_grade = np.dot(grades, weight)
print(final_grade)
```

```python
# convert to letter grade
letter = []
for i in final_grade:
    if i >= 90:
        letter.append('A')
    elif i >= 80:
        letter.append('B')
    elif i >= 70:
        letter.append('C')
    elif i >= 60:
        letter.append('D')
    else:
        letter.append('F')

print(letter)
```
